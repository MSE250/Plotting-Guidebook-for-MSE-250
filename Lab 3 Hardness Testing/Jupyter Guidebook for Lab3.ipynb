## 1. Import the modules and data

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool
import matplotlib.pyplot as plt
import pandas as pd

output_notebook()

filename = 'sample.csv'
dt = pd.read_csv(filename)

column = 1      # In Python, indexing starts from 0. So, when you refer to 1, it means the 2nd column.
start_row = 24  # The top header row is not counted when indexing. So, if you want to read from the 3rd row, you should input 1.
end_row = 33    # Reading will stop one row before the specified end_row.

# [24:33, 1] would mean reading the data from the 2nd column for rows starting from the 26th row up to the 34th row.
calibration_value = dt.iloc[start_row:end_row, column] 
measured_value = dt.iloc[start_row:end_row, column+1]

# The peculiar counting method can be quite confusing.
# So, it might be faster to just output the data directly and then make corrections.
print(calibration_value)
print(measured_value)
```

## 2. Plot graphs

Bokeh-produced charts allow for easy viewing of the values for each data point and support interactive features such as zooming and dragging.

```python
length = list(range(1, len(calibration_value) + 1))

p = figure(title="measured value vs calibration value", 
           x_axis_label='calibration value', 
           y_axis_label='measured value', 
           tools="pan,wheel_zoom,box_zoom,reset")

# The `p.line` function is used to draw a line chart.
# The `p.circle` is used to plot data points.
# If you remove the `p.line` function, you will be left with only the data points on the graph.
p.line(length, calibration_value, legend_label="calibration_value", line_width=2, color="blue")
p.circle(length, calibration_value, fill_color="white", size=8)

p.line(length, measured_value, legend_label="measured_value", line_width=2, color="green")
p.circle(length, measured_value, fill_color="green", size=8)

p.add_tools(HoverTool(
    tooltips=[
        ('#', '@x'),
        ('value', '@y')
    ]
))

# Add an auxiliary line
p.line(length, [-6*x+70 for x in length], legend_label="y = -6x+70", line_width=1.5, color="red", line_dash="dashed")

show(p)
```
